# Step 4
## Start Analysing

## Q1. How Many Matchs was Played In Every Vanue?

```python
x = df.groupby('venue')['match_id'].count().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(x['venue'], x['match_id'], marker='^', color='green')
plt.title('Count Of Match By Venue')
plt.xlabel('Venue')
plt.ylabel('Count Of Match')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

## Q2. How Many Match has Won By Every Team?

```python
x = df['match_winner'].value_counts()
sns.barplot(x=x.values,y=x.index)
plt.title('Number of Matches Won by Each Team in IPL 2022', fontsize=14)
plt.xlabel('Number of Matches Won', fontsize=12)
plt.ylabel('Team', fontsize=12)
```

## Q3. How Many Player Of The Match Won By Every Player?

```python
x= df['player_of_the_match'].value_counts()
sns.barplot(y=x.values,x=x.index)
plt.xticks(rotation=90)
plt.title('Count Of PTM For Player', fontsize=14)
plt.ylabel('PTM Count', fontsize=12)
plt.xlabel('Player', fontsize=12)
```

## Q4. Tose Win By Every Team..

```python
plt.xticks(rotation=45)
sns.countplot(x='toss_winner',data=df)
plt.title('Toss Winner Count', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Teams', fontsize=12)
```

## Q5. Average Score of first Inning After Losing Tose...

```python
df_filtered = df[(df['toss_winner'] != df['match_winner']) & 
                 (df['toss_decision'] == 'Field') ]
x = df_filtered.groupby('toss_winner')['first_ings_score'].mean().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(x['toss_winner'],x['first_ings_score'], marker='o', color='green')
plt.title('Average Score of first Inning After Losing Tose')
plt.xlabel('Team')
plt.ylabel('Average Scrore')
plt.grid(True)
plt.show()
```
